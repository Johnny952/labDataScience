Contentos con su desempe침o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a침o. Para esto se considerar치 que un alumno ha reprobado el a침o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena.

Cree una funci칩n que permita obtener dos Dataframes, el primero debe poseer solamente los alumnos que han reprobado en la prestigiosa universidad y el segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.

**To-Do:**

- [ ] Obtenga una m치scara DataFrame con los alumnos que reprobaron el a침o.
- [ ] Reemplaze todos los GPA reprobados con una **R**.

````python
def reprobados(df_grades):
    df_grades_ret = df_grades.copy()
    reprobados = df_grades_ret[df_grades_ret['GPA']<4]['names']
    df_grades_ret.loc[reprobados.index,'GPA']='R'
    return df_grades_ret,reprobados
````

Motivado por saber m치s sobre los datos, el Rector les solicita que genere una funci칩n que le permita comparar los promedios de diferentes **atributos categ칩ricos** del DataFrame. Para esto su funci칩n deber치 tener como entrada el DataFrame con todos los promedios (sin agregar las R's) y calcular el promedio **agrupando por etnias y el nivel educacional de los padres**.

**To-Do:**

- [ ] Comente lo que observa de los resultados.

````python
def promedios_agrupados(df_grades):
    p = df_grades.groupby(['race/ethnicity','parental level of education'])['race/ethnicity'].count()/10
    f = df_grades.groupby(['race/ethnicity','parental level of education']).mean()
    a = pd.concat([f,p],axis=1)
    a.rename(columns={'race/ethnicity': 'Percentage%'},inplace=True)
    #display(a)
    return a
````

```
La operaci칩n inner busca en los nombres estudiantes que est치n en ambos datasets y retorna uno nuevo con s칩lo estos nombres y sus columnas 
asociadas en ambos datasets, por lo que se pierden estudiantes que est치n en s칩lo uno de los datasets, en cambio la operaci칩n left utiliza

utiliza el dataset de la derecha como pivote, al cual se le van agregando valores del dataset de la izquierda. Por 칰ltimo la operaci칩n 
outer toma todos los nombres del dataset de la izquierda y la derecha, y a cada nombre le va asociando valores de cada uno de los dataset y 
los nombres que no tienen informaci칩n en uno de los dos dataset los rellena con None, pero 칠ste conserva en su totalidad los estudiantes 
de ambos datasets.

inner es la intersecci칩n de los conjuntos, left la uni칩n del dataset de la izquierda con la intersecci칩n, right la uni칩n del dataset de la
```
