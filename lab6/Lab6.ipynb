````python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
````

````python
try:
    from google.colab import drive
    # ejecutar en el caso de usar colab.
    drive.mount("/content/drive")
except Exception:
    pass
````

````python
df = pd.read_csv("./water_potability.csv").dropna()
X = df.drop(columns=["Potability"])
y = df.loc[:, ["Potability"]].values.ravel()
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, shuffle=True, stratify=y
)

ros = RandomOverSampler(random_state=42)
rus = RandomUnderSampler(random_state=42)
X_train, y_train = ros.fit_resample(X_train, y_train)
X_train, y_train = rus.fit_resample(X_train, y_train)


pipe = Pipeline(
    [("minmax", MinMaxScaler()), ("clf", RandomForestClassifier(random_state=42)),]
)

param_grid = {
    "clf__n_estimators": [200, 250, 300, 350],
    "clf__criterion": ["entropy", "gini"],
}

gs = GridSearchCV(pipe, param_grid, verbose=100, cv=10, n_jobs=-2)
gs.fit(X_train, y_train)

gs
````

````python
print(f"Mejor Modelo Grid Search: {gs.best_estimator_}\n")

print(f"Mejor Accuracy Grid Search: {gs.best_score_}\n")

best_random_forest = gs.best_estimator_
y_pred = best_random_forest.predict(X_test)

print("Reporte de Clasificaci√≥n\n")
print(classification_report(y_test, y_pred))
````

Y use `dump` de `joblib` para persistir el modelo reci√©n entrenado en el disco.

> ¬øQu√© es el modelo Cliente-Servidor?

        Es un modelo en el cual existen demandantes (clientes) y los proveedores de servicios o recursos (Servidor) en donde los demandantes realizan peticiones al servidor, y este les da una respuesta de forma as√≠ncrona, por lo que el cliente queda en espera mientras el servidor le responde.
    
> ¬øQu√© es una URI?
        
        Indentificador uniforme de recursos es una cadena de caracteres que se utiliza para identificar un recurso o nombre en internet, el cual permite la interacci√≥n entre diferentes recursos en internet.
    
    
> ¬øQu√© es HTTP?¬øQu√© significa que sea stateless?

        Un HTTP es un protocolo de transferencia de hipertexto, en donde se especifican las reglas (sintaxis, sem√°ntica y sincronizaci√≥n) de como se comunicar√°n 2 o m√°s entidades, como un computador, celular, entre otros. Todo esto es con el fin de transmitir la informaci√≥n por alg√∫n medio, lo cual se manifiesta con variaciones de alguna magnitud f√≠sica. Stateless significa que la solicitud actual no depende de ninguna de las solicitudes anteriores, por lo que es independiente.
    
> ¬øEn que consisten las operaciones de HTTP?: 
   
- GET
    
       Se refiere a la solicitud para pedir datos
        
- POST
    
        Sirve para enviar datos
        
- PUT
    
        Encargado de actualizar los datos
        
- DELETE
    
        Eliminar datos
        

> ¬øQu√© indican los c√≥digos de respuesta de HTTP?


       Indican distintas situaciones que ocurren a la hora de comunicarce con el servidor, por ejemplo si se pudo conectar con √©xito, si hubo alg√∫n error de conexi√≥n, si se encontr√≥ el archivo o no, entre otros.

> ¬øQu√© es un Hipermedio como HTML?

        Conjunto de procedimientos para escribir, dise√±ar y componer contenidos basados en texto, imagen, video, audio entre otros, de forma que los usuarios tengan la posibilidad de interactuar con ellos y viceversa.

> ¬øQu√© es una API?

        Es el conjunto de funciones que enmascaran otras funcionalidades m√°s complejas, de forma de dar simpleza al usuario y restringir el acceso a funcionalidades o datos sensibles.

> ¬øC√≥mo API REST engloba los t√©rminos anteriores?

        La Transferencia de estado representacional o representational state transfer (REST) consiste en un conjunto de principios en donde se encuentran: 
        -Un protocolo cliente/servidor sin estado( HTTP)
        -Las operaciones de HTTP como  POST, GET, PUT y DELETE
        -Una sintaxis universal para identificar los recursos
        -Uso de hipermedios

#### Consultas

Podemos hacer consultas, es decir, obtener algunos archivos seg√∫n alg√∫n criterio, a trav√©s un objeto `Query` y `search`:

````python
Mediciones = Query()

# Aqu√≠ Mediciones.Day representa a todas las llaves Day de todos los documentos.
# luego, el criterio de comparaci√≥n es el que indica que datos se traeran.
# en la pr√°ctica, trae 40 datos, del d√≠a 14 al 15.
db.search(Mediciones.Day <= 15)
````

Adem√°s, podemos mezclar distintos criterios de forma muy similar como lo hacemos en pandas:

En este caso, la siguiente consulta entrega todas las mediciones con ph entre 6 y 6.5 incluidas.

````python
Mediciones = Query()

db.search((Mediciones.ph >= 6) & (Mediciones.ph <= 6.5))
````

### Update

Podemos actualizar datos que hayan sido seleccionados usando una opraci√≥n y una `Query` como argumentos de `update`.

Las operaciones disponibles son:

    delete(key): delete a key from the document
    increment(key): increment the value of a key
    decrement(key): decrement the value of a key
    add(key, value): add value to the value of a key (also works for strings)
    subtract(key, value): subtract value from the value of a key
    set(key, value): set key to value


````python
from tinydb.operations import set

Mediciones = Query()

# en este caso, usaremos las mediciones del d√≠a 14 y fijaremos su ph a 7
db.update(set("ph", 7), Mediciones.Day == 14)
day=14
month=7
year=2021
db.update(set("ph", 5),((Mediciones.Day == day) & (Mediciones.Month == month)& (Mediciones.Year == year)))
# noten que esta funci√≥n retorna los indices de los documentos que fueron actualizados.
````

````python
db.search(Mediciones.Day == 14)
````

### Eliminado


Por √∫ltimo, podemos eliminar de la BBDD con una `Query` y la funci√≥n `remove`: 

````python
db.remove((Mediciones.Day == 15) & (Mediciones.Month == month)& (Mediciones.Year == year))
# noten que esta funci√≥n retorna los indices de los documentos que fueron borrados.
````

````python
db.search(Mediciones.Day == 14)
````
